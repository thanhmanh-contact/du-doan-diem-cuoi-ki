```python
import pandas as pd
import torch
from google.colab import drive

# Mount Google Drive
drive.mount("/content/drive")

# Đọc file Excel
path = "/content/drive/MyDrive/TRAIN2.xlsx"
df = pd.read_excel(path)

# Chuyển dữ liệu sang tensor
X = torch.tensor(df[["midterm"]].values, dtype=torch.float32)
Y = torch.tensor(df[["final"]].values, dtype=torch.float32)

# Khởi tạo weight và bias
W = torch.zeros((1, 1), dtype=torch.float32, requires_grad=True)
b = torch.zeros((1, 1), dtype=torch.float32, requires_grad=True)

# Hyperparameters
learning_rate = 0.01
epochs = 1000

# Training
for epoch in range(epochs):
    # Forward
    Y_pred = X @ W + b

    # Loss (MSE)
    loss = ((Y - Y_pred) ** 2).mean()

    # Backward
    loss.backward()

    # Update parameters
    with torch.no_grad():
        W -= learning_rate * W.grad
        b -= learning_rate * b.grad

        # Reset gradient ngay trong no_grad cho gọn
        W.grad.zero_()
        b.grad.zero_()

    # In loss mỗi 100 epoch để theo dõi
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}")

# Kết quả cuối
print("\nModel sau khi train:")
print(f"final = {W.item():.4f} * midterm + {b.item():.4f}")
```
